```r
# Function to run simulations with varying instrumental variables
set.seed(1024)

num_cores = detectCores() - 1
cl <- makeCluster(num_cores)
registerDoParallel(cl)
getDoParRegistered()
registerDoSNOW(cl)
strength_instr_list = seq(0.5,2,length.out=8)
results <- foreach(strength_instr = strength_instr_list,  .packages = c('dplyr', 'nprobust','MASS','caret')) %dopar% {
    paste("Running simulation with strength_instrument =", strength_instr)
      
    df_mc <- generate_synthetic_data(n = 10000, p = 5, p_instr = 2, gamma = 10, 
                                       propen_model = 'lr', strength_instrument = strength_instr, 
                                       heteroscedasticity = FALSE, mu1_type = 'non_linear')
      
    true_ate <- mean(df_mc$mu1 - df_mc$mu0)
      
  sim_results <- sim_eval_trial(
        true_ate = true_ate,
        num_trial = 500,
        n = 1000,
        p = 5,
        p_instr = 2,
        gamma = 10,
        propen_model = 'lr',
        strength_instrument = strength_instr,
        mu1_type = 'non_linear',
        ps_model = "lr",
        or_model = "lr",
        w_model = "nprobust", 
        heteroscedasticity = FALSE,
        random_seed = 1024,
        true_w = FALSE,
        df_mc = df_mc,
        jackknife = FALSE
    )
      
  return(list(
        strength_instr = strength_instr,
        mse = sim_results$mse,
        bias = sim_results$bias,
        results = sim_results$results  # Assuming sim_results has this field
    ))
  }
  
stopCluster(cl)

# Separate MSE, bias, and estimations results
mse_results <- lapply(results, function(res) data.frame(strength_instr = res$strength_instr, mse = res$mse))
bias_results <- lapply(results, function(res) data.frame(strength_instr = res$strength_instr, bias = res$bias))
estimations_results <- lapply(results, function(res) data.frame(strength_instr = res$strength_instr, estimations = res$results))

# Combine results into single data frames
mse_df <- do.call(rbind, mse_results)
bias_df <- do.call(rbind, bias_results)
estimations_df <- do.call(rbind, estimations_results)

```
